```python
# 1. Import the necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# 2. Define the filename
file_name = 'Cleaned Data.xlsx'

# 3. Load the Excel file into a DataFrame called 'df'
try:
    df = pd.read_excel(file_name)
    print("✅ File loaded successfully into the DataFrame!")

    # 4. Display the first 5 rows to confirm it looks right
    print("\nHere's a quick preview of your data:")
    display(df.head())

except FileNotFoundError:
    print(f"❌ Error: The file '{file_name}' was not found. Please re-upload it.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# Set a nice visual style for our plot
sns.set_style("whitegrid")

# Create the plot
plt.figure(figsize=(8, 6))
sns.countplot(data=df, x='gender', order=df['gender'].value_counts().index, palette='viridis')

# Add a title and labels to make it clear
plt.title('Gender Distribution of Applicants', fontsize=16)
plt.xlabel('Gender', fontsize=12)
plt.ylabel('Number of Applicants', fontsize=12)

# Display the plot
plt.show()
```

```python
# Create the Age Distribution plot (Histogram)
plt.figure(figsize=(10, 6))
sns.histplot(data=df, x='age', bins=30, kde=True, color='blue')

# Add a title and labels
plt.title('Age Distribution of Applicants', fontsize=16)
plt.xlabel('Age', fontsize=12)
plt.ylabel('Number of Applicants', fontsize=12)

# Display the plot
plt.show()
```

```python
# Get the top 10 countries by counting occurrences
top_10_countries = df['country'].value_counts().nlargest(10)

# Create the plot
plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_countries.values, y=top_10_countries.index, palette='plasma')

# Add a title and labels
plt.title('Top 10 Countries by Applicant Count', fontsize=16)
plt.xlabel('Number of Applicants', fontsize=12)
plt.ylabel('Country', fontsize=12)

# Display the plot
plt.show()
```

```python
# Get the top 10 institutions by counting occurrences
top_10_institutions = df['institution_name'].value_counts().nlargest(10)

# Create the plot
plt.figure(figsize=(12, 8))
sns.barplot(x=top_10_institutions.values, y=top_10_institutions.index, palette='cividis')

# Add a title and labels
plt.title('Top 10 Institutions by Applicant Count', fontsize=16)
plt.xlabel('Number of Applicants', fontsize=12)
plt.ylabel('Institution Name', fontsize=12)

# Display the plot
plt.show()
```

```python
# --- A More Creative and Convincing Visualization ---

# 1. Get the status counts and percentages
status_counts = df['status_description'].value_counts()
total_applicants = status_counts.sum()
status_percentages = (status_counts / total_applicants) * 100

# 2. Choose a color palette
colors = sns.color_palette('plasma', len(status_counts))

# 3. Create the Donut Chart
plt.figure(figsize=(10, 10))
plt.pie(status_counts, labels=status_counts.index, colors=colors,
        autopct='%1.1f%%', startangle=140, pctdistance=0.85,
        wedgeprops=dict(width=0.3, edgecolor='w'))

# 4. Add a circle in the center to make it a donut
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# 5. Add a title and ensure it's a perfect circle
plt.title('Applicant Status Funnel', fontsize=20, weight='bold')
plt.axis('equal')

# 6. Display the plot
plt.show()
```

```python
# --- A Polished, Client-Ready Visualization ---

# 1. Get the status counts
status_counts = df['status_description'].value_counts()

# 2. Create an 'explode' list. This will separate the largest slice.
# It makes a list of zeros, then sets the first element (the largest category) to 0.1
explode = [0.0] * len(status_counts)
explode[0] = 0.1  # Explode the 1st slice (the largest one)

# 3. Choose a color palette
colors = sns.color_palette('viridis', len(status_counts))

# 4. Create the Donut Chart
plt.figure(figsize=(12, 8))
wedges, texts, autotexts = plt.pie(status_counts,
                                   autopct='%1.1f%%',
                                   startangle=140,
                                   colors=colors,
                                   pctdistance=0.85,
                                   explode=explode,
                                   wedgeprops=dict(width=0.4, edgecolor='w'))

# 5. Add a circle in the center to make it a donut
centre_circle = plt.Circle((0,0),0.70,fc='white')
fig = plt.gcf()
fig.gca().add_artist(centre_circle)

# 6. Add a legend for clarity
plt.legend(wedges, status_counts.index,
          title="Application Status",
          loc="center left",
          bbox_to_anchor=(1, 0, 0.5, 1))

# 7. Add a title and ensure it's a perfect circle
plt.title('Applicant Status Funnel', fontsize=20, weight='bold')
plt.axis('equal')

# 8. Display the plot
plt.tight_layout() # Adjusts plot to ensure everything fits without overlapping
plt.show()
```

```python
# --- An Advanced, Insightful Visualization ---

# 1. Create the Violin Plot
plt.figure(figsize=(12, 8))
sns.violinplot(data=df, x='opportunity_category', y='age', palette='coolwarm')

# 2. Add titles and labels for clarity
plt.title('Age Distribution Across Opportunity Categories', fontsize=20, weight='bold')
plt.xlabel('Opportunity Category', fontsize=12)
plt.ylabel('Age of Applicant', fontsize=12)
plt.xticks(rotation=45) # Rotate category names for better readability

# 3. Display the plot
plt.tight_layout() # Adjusts plot to ensure everything fits
plt.show()
```

```python
# --- An Insightful Comparison Visualization ---

# 1. Create the grouped bar chart
plt.figure(figsize=(12, 8))
sns.countplot(data=df, y='opportunity_category', hue='gender',
              order=df['opportunity_category'].value_counts().index,
              palette='magma')

# 2. Add titles and labels for clarity
plt.title('Opportunity Preferences by Gender', fontsize=20, weight='bold')
plt.xlabel('Number of Applicants', fontsize=12)
plt.ylabel('Opportunity Category', fontsize=12)
plt.legend(title='Gender') # Add a legend to explain the colors

# 3. Display the plot
plt.tight_layout()
plt.show()
```

```python
# --- A Strategic Timeline Visualization ---

# 1. Ensure 'apply_date' is a datetime object
df['apply_date'] = pd.to_datetime(df['apply_date'])

# 2. Extract the month from the apply_date
# We create a new column representing the year and month
df['apply_month'] = df['apply_date'].dt.to_period('M')

# 3. Group by month and count the number of applications
monthly_applications = df['apply_month'].value_counts().sort_index()

# 4. Create the line chart
plt.figure(figsize=(14, 7))
monthly_applications.plot(kind='line', marker='o', linestyle='-', color='teal')

# 5. Add titles and labels for clarity
plt.title('Application Trends Over Time', fontsize=20, weight='bold')
plt.xlabel('Month', fontsize=12)
plt.ylabel('Number of Applications', fontsize=12)
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# 6. Display the plot
plt.tight_layout()
plt.show()
```
