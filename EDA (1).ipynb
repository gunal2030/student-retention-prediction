```python
print("=== Potential Data Quality Issues ===")



# 3. Constant columns (only 1 unique value â†’ no useful info)
constant_cols = [col for col in df.columns if df[col].nunique() <= 1]
print("\n Constant Columns:", constant_cols)

# 4. High cardinality categorical columns
cat_cols = df.select_dtypes(include=['object', 'category']).columns
for col in cat_cols:
    unique_ratio = df[col].nunique() / len(df)
    if unique_ratio > 0.9:  # >90% unique values = suspicious for categorical
        print(f"\n High cardinality column (possible ID): {col} ({df[col].nunique()} unique values)")

# 5. Outliers in numeric columns (using IQR rule)
num_cols = df.select_dtypes(include=['number']).columns
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df[(df[col] < (Q1 - 1.5 * IQR)) | (df[col] > (Q3 + 1.5 * IQR))]
    if not outliers.empty:
        print(f"\n Possible outliers detected in column: {col} ({len(outliers)} rows)")

# 6. Datetime columns with strange ranges
date_cols = df.select_dtypes(include=['datetime64[ns]']).columns
for col in date_cols:
    if df[col].notnull().sum() > 0:
        min_date, max_date = df[col].min(), df[col].max()
        print(f"\nðŸ”¹ Datetime column {col}: min={min_date}, max={max_date}")
```
